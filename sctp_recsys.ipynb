```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

X = df.drop(['ID_code', 'target'], axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(X_scaled)

def get_similar_customers(customer_index, n_neighbors=5):
    distances, indices = knn.kneighbors([X_scaled[customer_index]], n_neighbors=n_neighbors+1)
    similar_customers = indices[0][1:]
    similar_distances = distances[0][1:]
    return list(zip(similar_customers, similar_distances))

customer_id = 10
similar = get_similar_customers(customer_id)

print(f"Customers similar to customer {customer_id}:")
for cust, dist in similar:
    print(f"Customer {cust} with distance {dist:.4f}")
```
